<h1 style="text-align: center; color:Green;">Customer Shopping Behavior Analysis</h1>

<h2 style="color:Green;">Data Loading</h2>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.head()
```

```python
df.info()
```

<h2 style="color:Green;">Data Cleaning</h2>

```python
df.describe(include = 'all')
```

```python
df.isnull().sum()
```

- **Filling the Null Column**

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

- **Converting lower case columns**

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')

df  = df.rename(columns = {'purchase_amount_(usd)' : 'purchase_amount'})   # changing col name
```

```python
df.columns
```

<h2 style="color:Green;">Feature Engineering</h2>

- **Create a Column age_group**

```python
labels = ['Young Adult', 'Adult', 'Middle_aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

- **Create column purchase_frequency_days**

```python
frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Every 3 Months' : 90,
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days', 'frequency_of_purchases']].head(10)
```

```python
df[['discount_applied', 'promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

- **Removing promo_code_used**

```python
df= df.drop('promo_code_used', axis = 1)
df.columns
```

<h2 style="color:Green;">Exploratory Data Analysis (EDA) with Insights</h2>

- **Purchase Amount Distribution**

```python
sns.histplot(df['purchase_amount'], bins = 20)
plt.title('Distribution of Purchase Amount')
plt.show()
```

**Insight:**
- Most customers spend in a mid-range (around average)
- Very few high spenders (outliers)
- Like: most people buy normal items, few buy expensive stuff

- **Revenue by Category**

```python
df.groupby('category')['purchase_amount'].sum().plot(kind = 'bar')
plt.title('Revenue by Category')
plt.ylabel('Revenue')
plt.show()
```

**Insight:**
- One category (like Clothing) dominates revenue
- Helps business focus on top-performing products

- **Revenue by Gender**

```python
df.groupby('gender')['purchase_amount'].sum().plot(kind = 'pie', autopct = '%1.1f%%')
plt.title('Revenue by Gender')
plt.ylabel('')
plt.show()
```

**Insights:**
- Male customers contribute ~67.7% of total revenue
- Female customers contribute ~32.3% of total revenue
- Revenue from male customers is more than double that of female customers

- **Subscription vs Spending**

```python
sns.barplot(x='subscription_status', y='purchase_amount', data = df)
plt.title('Subscription vs Purchase Amount')
plt.show()
```

**Insight:**
- Average spending is similar for subscribers and non-subscribers
- Subscription does not significantly increase per-purchase value

- **Age Group vs Spending**

```python
sns.barplot(x='age_group', y='purchase_amount', data = df)
plt.title('Age Group vs Purchase Amount')
plt.show()
```

**Insight:**
- All age groups spend almost equally
- No strong age-based pricing or targeting pattern

- **Discount Impact on Spending**

```python
sns.barplot(x='discount_applied', y='purchase_amount', data = df)
plt.title('Discount vs Purchase Amount')
plt.show()
```

**Insight:**
- Discounts do not significantly increase average spending
- Customers may be purchasing similar-value items regardless of discount

- **Shipping Type vs Spending**

```python
sns.barplot(x='shipping_type', y='purchase_amount', data = df)
plt.title('Shipping Type vs Purchase Amount')
plt.xticks(rotation=45)
plt.show()
```

**Insight:**
- Shipping type has minimal impact on purchase amount
- Customers prioritize product over delivery option
